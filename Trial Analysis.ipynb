**Apakah Customer type : Member membeli lebih banyak barang?**

```python
from scipy.stats import mannwhitneyu
_, pval = mannwhitneyu(df.loc[df['Customer type']=='Member', 'Quantity'],
            df.loc[df['Customer type']=='Normal', 'Quantity'],
            alternative='greater')
if pval <= 0.05:
    print('Customer type: Member membeli banyak barang dibanding yang Normal')
else:
    print('Customer type: Member membeli barang dengan kuantitas yang sama dengan yang Normal')
```

**Apakah tiap Quantity memiliki range Unit Price yang berbeda?** 

```python
from scipy.stats import kruskal

kruskal(*[df[df['Quantity']==i]['Unit price'] for i in df['Quantity'].unique()])
```

Karena pvalue > 0.05, maka Ho diterima yaitu tiap quantity memiliki range unit price yang hampir sama

**Apakah range Unit Price tiap Product line sama atau berbeda?** 

```python
from scipy.stats import kruskal

kruskal(*[df[df['Product line']==i]['Unit price'] for i in df['Product line'].unique()])
```

Harga tiap Product line hampir mirip-mirip

**Apa hubungan Gender dengan pembelian Product Line tertentu?**

```python
prod_gen = pd.crosstab(index=df['Product line'], columns=df['Gender'], values=df['Quantity'], aggfunc='sum')
prod_gen.plot(kind='bar', figsize=(12,8))

plt.xticks(rotation=45)
plt.ylabel('Quantity')

for i in range(prod_gen.shape[0]):
    for j in range(prod_gen.shape[1]):
        if prod_gen.columns[j] == 'Female':
            plt.text(i-.2, prod_gen.values[i][j]+5, str(prod_gen.values[i][j]))
        else:
            plt.text(i+.05, prod_gen.values[i][j]+5, str(prod_gen.values[i][j]))

        
plt.legend(loc=9, prop={'size':12})

plt.savefig('Apa hubungan Gender dengan pembelian Product Line tertentu.png')

plt.show()
```

```python
prod_gp = pd.crosstab(index=df['Product line'], columns=df['Payment'], values=df['Quantity'], aggfunc='sum')[['Cash','Ewallet','Credit card']]

prod_gp.values

prod_gp.plot(kind='bar', figsize=(12,8))

plt.xticks(rotation=45)
plt.ylabel('Quantity')

for i in range(prod_gp.shape[0]):
    for j in range(prod_gp.shape[1]):
        if prod_gp.columns[j] == 'Cash':
            plt.text(i-.25, prod_gp.values[i][j]+5, str(prod_gp.values[i][j]))
        elif prod_gp.columns[j] == 'Ewallet':
            plt.text(i-.08, prod_gp.values[i][j]+5, str(prod_gp.values[i][j]))
        else:
            plt.text(i+.09, prod_gp.values[i][j]+5, str(prod_gp.values[i][j]))

plt.title('')

plt.legend(loc='upper right',  bbox_to_anchor=(1,1), prop={'size':12})

plt.savefig('Apa hubungan payment dengan pembelian Product Line tertentu.png')

plt.show()
```

Apabila kita melihat dengan seksama grafik diatas, ada beberapa hal yang signifikan seperti 
* untuk product line "Health and Beauty" ternyata lebih banyak dibeli oleh pria dibandingkan wanita.
* untuk product line "Sports and travel ternyata lebih banyak dibeli oleh wanita dibandingkan pria.

**Apakah ada hubugnan Waktu dengan pembelian di supermarket?**

```python
df['Time']
```

```python
k = (df['Time'].str.extract(r'([0-9]*)(?::)(?:[0-9]*)'))
df['Waktu'] = k[0].apply(lambda x: 'Subuh' if int(x)<6 else 'Pagi' if int(x)<12 else 'Sore' if int(x)<18 else 'Malam')
# df.loc[k['Waktu']=='Pagi']

df.groupby('Waktu', as_index=False)['Quantity'].sum().plot(kind='bar', x='Waktu', y='Quantity', figsize=(12,8), color=['blue','red','green'])
plt.xticks(rotation=45)

plt.ylabel('Quantity')
plt.show()
```

Lebih banyak orang membeli di waktu sore

```python
pd.crosstab(index=df['Waktu'], columns=df['Product line'], values=df['Quantity'], aggfunc='sum').plot(kind='bar', figsize=(20,10))

plt.xticks(rotation=45)
plt.ylabel('Quantity')

plt.legend(prop={'size':13})

plt.show()
```

* Sports and travel paling laku di waktu sore
* Home and lifestyle paling laku di waktu pagi
* food and beverages paling laku di waktu malam
